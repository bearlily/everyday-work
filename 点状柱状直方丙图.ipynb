```python
import matplotlib.pyplot as plt
x=[220,350,2304,3405,205]
y=[180,250,342,456,843]

plt.figure(figsize=(20,8),dpi=80)

plt.scatter (x,y)

plt.show
```

```python
movie_name=['','','','']
tickets=[456,345,455,231]

plt.figure(figsize=(20,8),dpi=80)

plt.bar(range(len(movie_name)),tickets,color=['b','r','g','m'])

plt.xticks(range(len(movie_name)),movie_name)

plt.title("")

plt.grid(linestyle="--",alpha=0.2)

plt.show()

```

```python
movie_name=['','','']

first_day=[100,200,300]
first_weekend=[200,300,50]

plt.bar(range(3),first_day,width=0.2,label="")
plt.bar([0.2,1.2,2.2],first_weekend,width=0.2,label="")

plt.legend()

plt.xticks([0.1,1.1,2.1],movie_name)

plt.show()
```

```python
import matplotlib.pyplot as plt

time=[133,244,355,466,577,476,536,100,761,590,453,893]
plt.figure(figsize=(20,8),dpi=80)

distance=150
group_num=int((max(time)-min(time))/distance)
plt.hist(time,bins=group_num,density=True)

plt.xticks(range(min(time),max(time),distance))

plt.grid(linestyle="--",alpha=0.5)


plt.show()
```

```python
import matplotlib.pyplot as plt

movie_name=['','','']
place_count=[100,200,300]

plt.figure(figsize=(20,8),dpi=80)

plt.pie(place_count,labels=movie_name,colors=['b','r','g'],autopct="%1.2f%%")

plt.legend()
plt.axis('equal')

plt.show()
```

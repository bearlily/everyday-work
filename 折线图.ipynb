```python
import random
import matplotlib.pyplot as plt
# matplotlib.rc("font",family='Microsoft YaHei')
x=range(60)
y_shanghai=[random.uniform(15,18) for i in x]
y_beijing=[random.uniform(1,3) for i in x]

plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y_shanghai,color="r",linestyle="--",label="")
plt.plot(x,y_beijing,color="m",linestyle="-.",label="")

plt.legend(loc=0)

x_label=["11{}".format(i) for i in x]
plt.yticks(range(0,40,5))
plt.xticks(x[::5],x_label[::5])

plt.grid(True,linestyle="--", alpha=0.5)

plt.xlabel("")
plt.ylabel("")
plt.title("")

plt.show()
```

```python
import random
import matplotlib.pyplot as plt
# matplotlib.rc("font",family='Microsoft YaHei')
x=range(60)
y_shanghai=[random.uniform(15,18) for i in x]
y_beijing=[random.uniform(1,3) for i in x]

figure, axes=plt.subplots(nrows=1,ncols=2,figsize=(20,8),dpi=80)

axes[0].plot(x,y_shanghai,color="r",linestyle="--",label="")
axes[1].plot(x,y_beijing,color="m",linestyle="-.",label="")

axes[0].legend(loc=0)
axes[1].legend(loc=0)


x_label=["11{}".format(i) for i in x]
axes[0].set_yticks(range(0,40,5))
axes[0].set_xticks(x[::10],x_label[::10])
axes[1].set_yticks(range(0,40,5))
axes[1].set_xticks(x[::10],x_label[::10])

axes[0].grid(True,linestyle="--", alpha=0.5)
axes[1].grid(True,linestyle="--", alpha=0.5)

axes[0].set_xlabel("")
axes[0].set_ylabel("")
axes[0].set_title("")
axes[1].set_xlabel("")
axes[1].set_ylabel("")
axes[1].set_title("")

plt.show()
```

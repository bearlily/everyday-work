```python
import random
import matplotlib.pyplot as plt
# matplotlib.rc("font",family='Microsoft YaHei')
x=range(60)
y_shanghai=[random.uniform(15,18) for i in x]
y_beijing=[random.uniform(1,3) for i in x]

plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y_shanghai,color="r",linestyle="--",label="上海")
plt.plot(x,y_beijing,color="m",linestyle="-.",label="北京")

plt.legend(loc=0)

x_label=["11点{}分".format(i) for i in x]
plt.yticks(range(0,40,5))
plt.xticks(x[::5],x_label[::5])

plt.grid(True,linestyle="--", alpha=0.5)

plt.xlabel("时间变化")
plt.ylabel("温度变化")
plt.title("上海、北京温度变化状况")

plt.show()
```

```python
import random
import matplotlib.pyplot as plt
# matplotlib.rc("font",family='Microsoft YaHei')
x=range(60)
y_shanghai=[random.uniform(15,18) for i in x]
y_beijing=[random.uniform(1,3) for i in x]

figure, axes=plt.subplots(nrows=1,ncols=2,figsize=(20,8),dpi=80)

axes[0].plot(x,y_shanghai,color="r",linestyle="--",label="上海")
axes[1].plot(x,y_beijing,color="m",linestyle="-.",label="北京")

axes[0].legend(loc=0)
axes[1].legend(loc=0)


x_label=["11点{}分".format(i) for i in x]
axes[0].set_yticks(range(0,40,5))
axes[0].set_xticks(x[::10],x_label[::10])
axes[1].set_yticks(range(0,40,5))
axes[1].set_xticks(x[::10],x_label[::10])

axes[0].grid(True,linestyle="--", alpha=0.5)
axes[1].grid(True,linestyle="--", alpha=0.5)

axes[0].set_xlabel("时间变化")
axes[0].set_ylabel("温度变化")
axes[0].set_title("上海温度变化状况")
axes[1].set_xlabel("时间变化")
axes[1].set_ylabel("温度变化")
axes[1].set_title("北京温度变化状况")

plt.show()
```

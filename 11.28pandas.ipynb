```python
import pandas as pd
import numpy as np
from copy import deepcopy
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

df=pd.read_excel(r"C:\Users\Andy\Desktop\超市营业额2.xlsx")
df.drop_duplicates(inplace=True)
for i in df[df.交易额.isnull()].index:
    df.loc[i,'交易额']=round(df.loc[df.姓名==[i,'姓名'],'交易额'].median())
df.loc[df.交易额<500,'交易额']=500
df.loc[df.交易额>3000,'交易额']=3000

display(df)
```

```python
df2=df.groupby(by="日期",as_index=False).agg({"交易额":"sum"}).nsmallest(3,"交易额")
print(pd.to_datetime(df2.日期).dt.weekday)
```

```python
df3=deepcopy(df)
df3['工号']=df['工号'].map(lambda s:str(s)[-1]+str(s))
df3.to_excel()
```

```python
writer=pd.ExcelWriter(r"C:\Users\Andy\Desktop\各员工数据.xlsx")
names=set(df['姓名'].values)
for name in names:
    dff=df[df.姓名==name]
    dff.to_excel(writer,sheet_name=name,index=False)
writer.close()
```

```python
df8=df.loc[:,
           ['日期','柜台','交易额']].groupby(by=['日期','柜台'],as_index=False).sum()
df8=df8.pivot(index="日期",columns="柜台",values="交易额")
df8.plot()
myfont = fm.FontProperties(fname=r'c:\Windows\Fonts\sTKAITI.ttf')
plt.xlabel('日期',fontproperties='stkaiti')
plt.xticks(rotation=10)
plt.legend(prop=myfont)
plt.savefig(r"C:\Users\Andy\Desktop\1.png",dpi=200)
```

```python
plt.rcParams['font.sans-serif'] = ['simhei']
df9=df.loc[:,['柜台','交易额']].groupby(by='柜台',as_index=False).sum()
df9.plot(x='柜台',y='交易额',kind='pie',labels=df.柜台.values)
myfont =fm.FontProperties(fname=r'c:\Windows\Fonts\sTKAITI.ttf')
plt.legend()
plt.savefig(r"C:\Users\Andy\Desktop\2.png",dpi=200)
```

```python
df10=df.groupby(by=['姓名','日期'],as_index=False).sum()
df10=df10.pivot(index="姓名",columns="日期",values="交易额")
df10=df10[df10.sum(axis=1)<5000][:5]
with open(r"C:\Users\Andy\Desktop\业绩差员工.txt",'w',encoding='utf-8') as fp:
    for name in df10.index.values:
        gonghao=df[df.姓名==name].工号.values[0]
        fp.write(str(gonghao)+','+name+'\n')
```

```python
df11=pd.crosstab(df.姓名,df.柜台,df.交易额, aggfunc="sum").apply(round)
df11.plot(kind="bar")
font = fm.FontProperties(fname=r'c:\Windows\Fonts\sTKAITI.ttf')
plt.xlabel('员工业绩分布',fontproperties='simhei')
plt.xticks(fontproperties='simhei')
plt.legend(prop=font)
plt.savefig(r"C:\Users\Andy\Desktop\3.png",dpi=200)
```
